```python
"""Aparies frequentes dos sorteios"""

import pandas as pd
import matplotlib.pyplot as plt

 
df = pd.read_csv("https://raw.githubusercontent.com/cardosource/OraculoAI/main/2023.csv", header=None)
aparicoes = {i: [0] * 6 for i in range(1, 61)}

 
for _, row in df.iterrows():
    for i, num in enumerate(row):
        if num in range(1, 61):
            aparicoes[num][i] += 1

 
porcentagens = {}
for i in range(1, 61):
    total = sum(aparicoes[i])
    if total != 0:
        porcentagens[i] = [x / total * 100 for x in aparicoes[i]]
    else:
        porcentagens[i] = [0] * 6

fig, ax = plt.subplots(figsize=(12, 8))
for i in range(1, 61):
    ax.bar(range(6), porcentagens[i], bottom=sum(porcentagens[j][k] for j in range(1, i) for k in range(6)))
plt.legend(range(1, 61))
plt.xticks(range(6), ["1", "2", "3", "4", "5", "6"])
plt.xlabel("Posio")
plt.ylabel("Porcentagem de aparies")
plt.show()

 
frequencias = {}
for i in range(1, 61):
    frequencias[i] = [j for j, x in enumerate(aparicoes[i]) if x == max(aparicoes[i])]
```
